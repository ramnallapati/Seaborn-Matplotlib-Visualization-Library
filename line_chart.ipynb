```python
# convert month category to date and time

df["Month"] = pd.to_datetime(df["Month"])

df.head()
```

```python
from datetime import datetime
```

```python
df["Year"] = df["Month"].dt.year
```

```python
df['month'] = df["Month"].dt.month_name()
```

```python
df.drop(columns=["Month"],axis=1,inplace=True)
df.head()
```

```python
# Basic syntax

sns.lineplot(x="Year",y="Passengers",data=df)
plt.title("Passengers vs Years")
plt.show()
```

```python
sns.lineplot(x="month",y="Passengers",hue="Year",data=df)
plt.title("Passengers in Months vs Years")
plt.show()
```

```python
# changing line styles

sns.lineplot(x="month",y="Passengers",style="Year",data=df)
plt.title("Passengers in Months vs Years")
plt.show()
```

```python
# add markers on points

sns.lineplot(x="month",y="Passengers",data=df,marker="o")
plt.title("Months vs Passengers")
plt.show()
```

```python
# changing the estimator sum

sns.lineplot(x="Year",y="Passengers",data=df,marker='o',estimator=np.sum)
plt.title("Years vs Passengers")
plt.show()
```
