<h3 style="color:orange;text-align:center">Area Charts </h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1)

days = list(range(1,101))

df = pd.DataFrame({
  "days":days,
  "Product_A":np.random.randint(50,150,100).cumsum(),
  "Product_B":np.random.randint(30,120,100).cumsum(),
  "Product_C":np.random.randint(20,100,100).cumsum()
})

df.head()
```

```python
# simple area chart

sns.lineplot(x="days",y="Product_A",data=df)

plt.fill_between(df['days'],df['Product_A'],alpha=0.3)
plt.show()
```

```python
# Plotting Multiple area charts

# line plots

sns.lineplot(x="days",y="Product_A",data=df,label="Product A")
sns.lineplot(x="days",y="Product_B",data=df,label="Product B")
sns.lineplot(x="days",y="Product_C",data=df,label="Product C")

# Fill Areas
plt.fill_between(df['days'],df['Product_A'],alpha=0.2)
plt.fill_between(df['days'],df['Product_B'],alpha=0.2)
plt.fill_between(df['days'],df['Product_C'],alpha=0.2)


plt.title("Area Chart of Product Sales over Time")
plt.xlabel("Day")
plt.ylabel("Sales")

plt.legend(loc="upper left")
plt.show()
```

```python
# Stacked Area chart

plt.stackplot(
  df["days"],
  df["Product_A"],
  df["Product_B"],
  df['Product_C'],
  labels=["A","B","C"]
)

plt.legend(loc="upper left")
plt.title("Stacked Area Chart")
plt.show()
```
