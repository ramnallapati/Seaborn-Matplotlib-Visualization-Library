<h3 style="color:blue;text-align:center">Pie Charts </h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "User_ID": range(1, 1001),
    "Country": np.random.choice(
        ["India", "USA", "UK", "Germany", "Canada", "Australia"],
        1000
    ),
    "Device": np.random.choice(
        ["Mobile", "Laptop", "Tablet"],
        1000
    ),
    "Subscription": np.random.choice(
        ["Free", "Basic", "Premium"],
        1000
    ),
    "Gender": np.random.choice(
        ["Male", "Female", "Other"],
        1000
    )
})

print(df.head())
```

```python
df['Country'].value_counts().plot(kind='pie')
```

```python
# subscription distribution

count_device = df['Device'].value_counts()

plt.pie(count_device,labels=count_device.index,autopct="%1.2f%%")

plt.title('Device Types')
plt.show()
```

```python
subscription_counts = df['Subscription'].value_counts()

plt.pie(subscription_counts,labels=subscription_counts.index,autopct='%1.1f%%',shadow=True,startangle=140)

plt.title('Subscription Status')
plt.show()
```
