```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3 style="color:orange;text-align:center">Bar Charts </h3>

```python
data = {
    "Product": ["Laptop", "Laptop", "Phone", "Phone", "Tablet", "Tablet"],
    "Sales": [200, 250, 300, 280, 150, 170],
    "Region": ["North", "South", "North", "South", "North", "South"]
}

df = pd.DataFrame(data)

df.head()
```

```python
sns.barplot(x="Product",y="Sales",data=df)
plt.title("Sales vs Products in Mean")
plt.show()

# mean of each sale
```

```python
# finding the product sales in sum

sns.barplot(x="Product",y="Sales",data=df,estimator=np.sum)
plt.title("Product vs Sales in Sum")
plt.show()
```

```python
# changing confidence intervals


sns.barplot(x="Product",y="Sales",data=df,ci=80)
plt.title("Product vs Sales in 95 CI")
plt.show()
```

```python
# used to compare sub-categories
sns.barplot(x="Product",y="Sales",data=df,hue="Region")
plt.title("Product sales vs Regions")
plt.show()
```

```python
# order of Bars

sns.barplot(x="Product",y="Sales",order=["Tablet","Laptop"],data=df)
plt.show()
```

```python
# color Customization

sns.barplot(x="Product",y="Sales",hue="Region",data=df,palette="viridis")
plt.show()
```

```python
# cat plot (Figure-level)

sns.catplot(x="Product",y="Sales",col="Region",kind="bar",data=df,palette="viridis")
plt.show()
```
