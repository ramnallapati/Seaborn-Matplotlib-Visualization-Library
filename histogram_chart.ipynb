````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
np.random.seed(0)

df = pd.DataFrame({
  "age":np.random.randint(18,60,1000),
  "salary":np.random.randint(5000,15000,1000),
  "experience":np.random.randint(0,35,1000),
  "department":np.random.choice(["IT","HR","Sales",'Finance'],1000),
  "gender":np.random.choice(["Male","Female"])
})

df.head()
````

````python
# basic syntax

sns.histplot(x="salary",data=df)
plt.show()
````

````python
# Histogram with bins

sns.histplot(data=df,x="salary",bins=20)
plt.title("Salary Distribution")
plt.show()
````

````python
# Histogram with Kde

sns.histplot(data=df,x="salary",bins=20,kde=True)
plt.title("Salary Distribution")
plt.show()
````

```
Stat
1. Count
2. Density
3. Probability
```

````python
# multiple category comparision

sns.histplot(data=df,x="salary",hue="department",bins=20,palette="viridis",stat="probability")
plt.title("Salary Distribution in Gender")
plt.show()
````

````python
# multiple columns Histogram

sns.histplot(data=df[["salary","experience"]],bins=20)
plt.show()
````
