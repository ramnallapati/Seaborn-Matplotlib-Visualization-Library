<h3 style="color:blue;text-align:center">Box-Whisker Plot</h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
  "Age":np.random.normal(10,80,100),
  "Salary":np.random.normal(50000,120000,100),
  "Experience":np.random.normal(5,2,100),
  "Department":np.random.choice(["IT","HR",'Finance',"Sales"],100)
})

# adding outliers
outliers = pd.DataFrame({
    "Department": ["IT"] * 10,
    "Salary": np.random.normal(150000, 20000, 10),
    "Experience": np.random.normal(15, 3, 10),
    "Age": np.random.normal(50, 5, 10)
})

df = pd.concat([df,outliers],ignore_index=True)
df.head()
```

```python
sns.boxplot(x=df["Age"])
plt.show()
```

```python
# group-wise box plot

sns.boxplot(x="Department",y="Age",data=df)
plt.show()
```

```python
# group-wise Horizontal box plot

sns.boxplot(x="Age",y="Department",data=df)
plt.show()
```

```python
# adds extra grouping

sns.boxenplot(x="Department",y='Age',hue="Department",data=df)
plt.show()
```

```python
# swarm plot

sns.swarmplot(x="Department",y="Salary",data=df)
sns.boxplot(x="Department",y="Salary",data=df)
plt.show()
```

```python
# Log scale (for skewed Data)

plt.yscale("log")
sns.boxplot(x="Department",y="Salary",data=df)
plt.show()
```
