```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

df = pd.DataFrame({
    'Math': np.random.randint(40, 100, 50),
    'Physics': np.random.randint(35, 100, 50),
    'Chemistry': np.random.randint(30, 100, 50),
    'Biology': np.random.randint(45, 100, 50),
    'English': np.random.randint(50, 100, 50),
    'History': np.random.randint(40, 95, 50),
    'Geography': np.random.randint(42, 98, 50),
    'Computer': np.random.randint(60, 100, 50)
})


df.head()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidth=0.5)
plt.title("Subject Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
  df.corr(),
  annot=True,
  cmap="RdBu",
  fmt=".2f",
  linewidth=1,
  linecolor='black',
  cbar=True
)

plt.title("Correlation Heatmap")
plt.show()
```
