<h3 style="color:blue;text-align:center">Scatter Plot</h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
  "age":np.random.randint(18,70,100),
  "income":np.random.randint(20000,100000,100),
  "gender":np.random.choice(["Male","Female"],100),
  "experience":np.random.randint(1,15,100)
})

df.head()
```

```python
# basic scatter plot

sns.scatterplot(x="age",y="experience",data=df)
plt.title("Age vs Experience")
plt.show()
```

```python
# compare groups
sns.scatterplot(x="age",y="income",hue="gender",data=df)
plt.title("Distribution of age and income in Gender")
plt.show()
```

```python
# variation with sizes

sns.scatterplot(x="age",y="income",hue="gender",size="experience",data=df)
plt.title("Distribution of age and income in Gender")
plt.show()
```

```python
# adding palette and alpha (Transparency)

sns.scatterplot(x='age',y='income',hue="gender",size="experience",palette="coolwarm",data=df)
plt.show()
```

```python
# regression line in scatter plot

sns.regplot(x="income",y="experience",data=df)
plt.show()
```

```python
# Facet (Multiple Plots)

sns.relplot(x="age",y="income",col="gender",data=df)
plt.show()
```
