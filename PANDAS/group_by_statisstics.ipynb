<h3 style="color:orange;text-align:center">Groupby Methods [Huge Categories] </h3>

1. agg()
2. aggregate()
3. sum()
4. mean()
5. median()
6. min()
7. max()
8. count()
9. size()
10. std()
11. var()
12. sem()
13. prod()
14. first()
15. last()
16. nth()
17. transform()
18. filter()
19. ngroup()
20. cumcount()

````python
df = pd.DataFrame({
  "Department":np.random.choice(['IT','HR','Sales'],9),
  "Employee":['A','B','C','D','E','F','G','H','I'],
  'Salary':[60000,70000,50000,52000,45000,47000,65000,51000,48000],
  'Experience':[5,7,3,4,2,3,6,4,3]
})

print(df)
````

````python
# groupby() -> split-apply-combine

df.groupby('Department')
````

````python
# sum

# -> total salary per department

df.groupby('Department')['Salary'].sum()
````

````python
# mean

# -> average salary per department

df.groupby('Department')['Salary'].mean().round()
````

````python
# medain

# -> middel values

df.groupby('Department')['Salary'].median()
````

````python
# min,max

# -> minimum and maximum values in each department

print(df.groupby('Department')['Salary'].min())
print(df.groupby('Department')['Salary'].max())
````

````python
# count()

# -> counts non-null values only

df.groupby('Department')['Salary'].count()
````

````python
# size()

# -> count all rows including null values

df.groupby('Department')['Salary'].size()
````

````python
# std -> standard deviation

df.groupby('Department')['Salary'].std()
````

````python
# var -> variance

df.groupby('Department')['Salary'].var()
````

````python
# sem -> Standard Error

df.groupby('Department')['Salary'].sem()
````

````python
# prod() -> Product

df.groupby('Department')['Experience'].prod()
````

````python
# first(),last() rows in each group

print(df.groupby('Department').first())

print(df.groupby('Department').last())

# nth row
print(df.groupby('Department').nth(2))
````

```
agg() / aggregate()
```

````python
df.groupby('Department')['Salary'].agg(['sum','mean','max']).round()
````

````python
print(df.groupby('Department').agg({
  'Salary':['sum','mean'],
  'Experience':['max','min']
}).round())
````

```
transform()
```
1. Returns same size output as original DataFrame

````python
df['Avg_salary'] = df.groupby('Department')['Salary'].transform('mean').round()

print(df)
````

```
filter()
```
1. Filters groups based on condition

````python
df.groupby('Department').filter(lambda x:x['Salary'].mean() > 50000)
````

```
ngroup()
```
1. Assigns group numbers

````python
df['Group_ID'] = df.groupby('Department').ngroup()

print(df)
````

```
cumcount()
```
1. Counts occurences within group

````python
df['Count'] = df.groupby('Department').cumcount()

print(df)
````

```
pivot
```

1. Does not Handles Duplicate


````python
# # Methods
# 1. index
# 2. columns
# 3. values
# 4. aggfun
````

````python
df = pd.DataFrame({
  'Date':['2024-01','2024-01','2024-02','2024-02'],
  'Product':['A','B','A','B'],
  'Sales':[100,200,150,250]
})

print(df)
````

```
pivot_table
```

1. Handles Duplicate values

````python
print(df.pivot_table(index='Date',columns='Product',values='Sales',aggfunc='sum'))
````

```
melt
```

1. Opposite of pivot

````python
# id_vars -> columns keep it as unchanged
# value_vars -> columns you want to convert into rows
# var_name -> Name of the column
# value_name -> Name for the vlaues column
````

````python
df = pd.DataFrame({
  'Name':['Ram','Shyam'],
  'Math':[90,80],
  'Science':[85,75]
})

result = pd.melt(
  df,
  id_vars=['Name'],
  value_vars=['Math','Science'],
  var_name='Subject',
  value_name='Marks'
)

print(result)
````

<h3 style='color:pink;text-align:center'>Time Series Methods </h3>

1. to_datetime()
2. to_timedelta()
3. date_range()
4. resample()
5. asfreq()
6. shift()
7. tshift()
8. diff()
9. tz_localize()
10. tz_convert()

````python
df = pd.DataFrame({
  'Date':pd.date_range('2024-01-01',periods=5,freq='D'),
  'Sales':[100,150,200,130,170]
})

print(df)
````

```
to_datetime() -> convert string to datetime
```

````python
df['Date'] = pd.to_datetime(df['Date'])
print(df)
````

```
to_timedelta() -> converts time diffrences
```

````python
# Units
# 1. D = Days
# 2. H = Hours
# 3. m = Minutes
# 4. s = Seconds
# 5. ms = milliseconds
# 6. us = Microseconds
# 7. ns = Nanoseconds
````

````python
df['days_added'] = pd.to_timedelta(df['Sales'],unit='D')

df['Shipping_date'] = df['Date'] + df['days_added']
print(df)
````

```
resample
```

1. Change time frequency

````python
dff = df.copy()

dff = dff.set_index('Date')

print(dff)
````

````python
# Daily -> Monthly sales

print(dff.resample('ME').agg({
  'Sales':'sum'
}))
````

````python
# shift
# -> shift data up/down

dff['prev_sales'] = dff['Sales'].shift(1)

print(dff)
````

````python
# diff() => Diffrence between rows

dff['daily_change'] = dff['Sales'].diff()

print(dff)
````

````python
# convert timezone

dff = dff.tz_localize('Asia/Kolkata')

print(dff)
````

<h3 style="color:orange;text-align:center">Mathematical Methods</h3>

````python
df = pd.DataFrame({
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Laptop', 'Mobile'],
    'Price': [50000, 20000, 15000, 52000, 21000],
    'Quantity': [1, 2, 3, 1, 2],
    'Discount': [5000, 2000, 1500, 6000, 2500],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics']
})

print(df)
````

````python
# add

df['Total'] = df['Price'].add(df['Quantity'],fill_value=0)

print(df)
````

````python
# subtraction

df['Net'] = df['Price'].sub(df['Discount'])

print(df)
````

````python
# Multiplication

df['Revenue'] = df['Price'].mul(df['Quantity'])

print(df)
````

````python
# division

df['price_per_item'] = df['Price'].div(df['Quantity'])

print(df)
````

````python
# power

df['Square'] = df['Quantity'].pow(2)

print(df)
````

````python
# absolute value

df['diff'] = (df['Price']-df['Discount']).abs()

print(df)
````

````python
# Limit values

df['Price_limit'] = df['Price'].clip(lower=2000,upper=5000)

print(df)
````

````python
# eval -> perform expression

df.eval('Total_price = Price*Quantity - Discount',inplace=True)

print(df)
````

````python
# where
# -> Keep values if condition true, else replace

# IF condition TRUE → keep value  
# ELSE → replace

df['Price_where'] = df['Price'].where(df['Price']>20000,20000)
print(df)
````

````python
# mask()
# Opposite of where

# IF condition TRUE → keep value  
# ELSE → replace

df['Price_mask'] = df['Price'].mask(df['Price']>20000,50000)

print(df)
````

````python
# explode()
# Expand list into rows

df2 = pd.DataFrame({
  'Product':['Laptop','Mobile'],
  'Tags':[['Tech','Portable'],['Smart','Android']]
})


print(df2.explode('Tags'))
````

````python
# find duplicates

df.duplicated()
````

````python
# Remove duplicates

print(df.drop_duplicates())
````

````python
# drop()
# Remove rows/columns

df.drop('Category',axis=1,inplace=True)

print(df)
````

<h3 style="color:blue;text-align:center">Categorical Methods</h3>

````python
df = pd.DataFrame({
    'customer': ['Ram', 'Shyam', 'Anil', 'Ram', 'Anil'],
    'city': ['Hyderabad', 'Delhi', 'Mumbai', 'Hyderabad', 'Delhi'],
    'membership': ['Gold', 'Silver', 'Gold', 'Platinum', 'Silver'],
    'age': [25, 30, 35, 28, 40],
    'spending': [1000, 1500, 2000, 1200, 1800]
})

print(df)
````

````python
df.dtypes
````

````python
# Get Unique categories

df['membership'].cat.categories
````

````python
# Rename category lables

df['membership'] = df['membership'].cat.rename_categories({
  'Gold':'G',
  'Silver':'S',
  'Platinum':'P'
})

print(df)
````
