````python
import numpy as np
import pandas as pd
import seaborn as sns
import pandas as pd
````

```
Creating Series
```

````python
# Series

data = [1,2,3,4,5,6,7]

pd.Series(data,index=None)
````

````python
s = pd.Series([10,20,30],index=["a","b","c"])
print(s)
````

```
DataFrame
```

````python
df = pd.DataFrame({
  "Name":["Ram","Sam"],
  "Age":[25,30]
})

print(df)
````

````python
# Index

df = pd.DataFrame({
  "Name":["Ram","Sam"],
  "Age":[25,30]
})

df.index = [7,8]

print(df)
````

````python
# MultiIndex

arrays = [["A","A","B"],[1,2,1]]

mi = pd.MultiIndex.from_arrays(arrays)

s = pd.Series([10,20,30],index=mi)

print(s)
````

````python
## Categorical

cat = pd.Categorical(["low","medium","high","low"])

print(cat)
````

````python
# array

arr = pd.array([1,2,None])
print(arr)
````

````python
# date_range

dates = pd.date_range("2024-01-01",periods=10)

print(dates)
````

````python
# period range for Months,freq

pr = pd.period_range("2024-01-01",periods=3,freq="M")

print(pr)
````

````python
# Interval range

pd.interval_range(start=0,end=10)
````
