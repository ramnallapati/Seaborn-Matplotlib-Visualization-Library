<h3 style="color:orange;text-align:center">Transformation Methods</h3>

````python
# astype()
# - Convert data type of a column or entire dataframe

df = pd.DataFrame({
  "Age":['2','3','4']
})

print(df.dtypes)
df.head()
````

````python
df = df.astype(int)
df.dtypes
````

````python
# converting data type of multiple columns

df = df.astype({"Doors":"float","Owner_Count":"float"})

df.dtypes
````

````python
# apply()
# - Apply a function along rows or columns


df = pd.DataFrame({
  "A":[1,2,3]
})

df['A_Squared'] = df['A'].apply(lambda x : x**2)

df.head()
````

````python
# applymap()
# - Apply function to every element of DataFrame

df = pd.DataFrame({
  "A":[1,2],
  "B":[3,4]
})

df.applymap(lambda x: x*10)
````

````python
# map()
# - Transform values in a Series (Single column)

df = pd.DataFrame({
  "gender":np.random.choice(["Male","Female"],10)
})

df["gender"] = df["gender"].map({"Male":0,"Female":1})

df.head()
````

````python
# transform()
# - Transform Data but keep same shape


df = pd.DataFrame({
  "A":[1,2,3]
})

df['A_doubled'] =df['A'].transform(lambda x: x*2)
df
````

````python
# pipe()
# - Chain Functions for clean and readable code


def add_column(df):
  df["Sum"] = df["A"] + df["A_doubled"]
  return df

df = df.pipe(add_column)

df
````

<h3 style="color:orange;text-align:center">Sorting Methods</h3>

```
sort_values()

1. by -> column names
2. axis -> axes
3. ascending
4. inplace
5. na_position -> First/Last
```

````python
df = pd.DataFrame({
  "Age":np.random.randint(10,60,20),
  "gender":np.random.choice(["Male","Female","Other"],20),
  "department":np.random.choice(["HR","Developer","Team Lead","UI/UX"],20)
})
````

````python
# sort values
df.sort_values(by="Age",axis=0,ascending=True,na_position="first")
````

````python
df
````

````python
df
````

````python
dff = pd.DataFrame({
  "Name":["Ram","Prasanth","Bala Murali","Lakshmi","Malleswari"],
  "Gender":np.random.choice(["Male","Female","Other"],5),
  "Department":np.random.choice(["HR","Finance","Tech","Developer"],5),
  "Age":np.random.randint(23,35,5)
})

print(dff.head())
````

````python
# setting indexes

dff.index = ['a','b','c','d','e']

print(dff.head())
````

````python
# setting column name as a index

dff = dff.set_index("Name")

print(dff.head())
````

````python
# sort_index()
# - It is used to sort DataFrame or series based on index (row labels,or column labels)

dff = dff.sort_index(ascending=True)

print(dff)
````

````python
# sort columns instead of rows

dff.sort_index(axis=1,inplace=True)

print(dff)
````

````python
# resetting index means back to normal

dff = dff.reset_index()

print(dff.head())
````

<h3 style="color:orange;text-align:center">Merging and Joining</h3>

```
merge()
```

* Methods
1. inner -> common values only
2. left -> all from left
3. right -> all from right
4. outer -> all from both

````python
orders = pd.DataFrame({
  "order_id":[101,102,103,104],
  "customer_id":[1,2,2,5],
  "amount":[500,700,300,900]
})

print(orders)
````

````python
orders = orders.set_index("customer_id")
print(orders)
````

````python
customers = pd.DataFrame({
  "customer_id":[1,2,3,4],
  "name":["Ram","Shyam","Ravi","Anu"],
  "city":["Hyderabad","Delhi","Mumbai","Chennai"]
})

print(customers)
````

````python
payments = pd.DataFrame({
  "order_id":[101,102,103],
  "status":["Paid","Pending","Paid"]
})

print(payments)
````

````python
backup_customers = pd.DataFrame({
  "customer_id":[1,2,3,4],
  "name":["Ram",None,"Ravi","Anu"],
  "city":["Hyderabad","Delhi",None,"Chennai"]
})

print(backup_customers)
````

````python
# combine customers + orders

pd.merge(customers,orders,on="customer_id",how="inner")
````
